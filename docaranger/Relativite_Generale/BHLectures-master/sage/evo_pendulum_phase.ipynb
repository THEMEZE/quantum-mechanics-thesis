# Pendulum phase space

```python
%display latex
```

```python
x, y = var('x y')
H = y^2 + 1 - cos(x)
H
```

```python
ym = 2
graph = contour_plot(H, (x, -pi, pi), (y, -ym, ym), 
                    contours=12, fill=False, cmap=['grey'], 
                    plot_points=200,
                    axes_labels=[r'$\theta$', r'$p_\theta$'])
graph
```

```python
Ha = 1
Hb = 3
graph += implicit_plot(H==Ha, (x, -pi, pi), (y, -ym, ym), 
                      color='red', linewidth=2) \
        + implicit_plot(H==Hb, (x, -pi, pi), (y, 0, ym), 
                      color='blue', linewidth=2)
graph
```

```python
graph += plot_vector_field((2*y, -sin(x)), (x, -pi, pi), (y, -ym, ym),
                           color='grey')
graph
```

```python
xa = -1
ya = sqrt(Ha - 1 + cos(xa))
xb = -2.6
yb = sqrt(Hb - 1 + cos(xb))
graph += point((xa, ya), color='red', size=60) \
         + point((xb, yb), color='blue', size=60) \
         + text(r'$a$', (xa, 1.4*ya), fontsize=28, color='red') \
         + text(r'$b$', (xb, 1.25*yb), fontsize=28, color='blue')
graph
```
