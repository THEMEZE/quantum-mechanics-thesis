```python
# Exemple de code Python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Exemple de plot pour l\'idée 1')
plt.show()
```

```python
# Exemple de code Julia
using Plots

x = 0:0.1:10
y = sin.(x)

plot(x, y, title="Exemple de plot pour l'idée 1")
```
