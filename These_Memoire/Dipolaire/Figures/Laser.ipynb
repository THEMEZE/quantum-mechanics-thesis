```python
import numpy as np

# Constantes physiques
hbar = 1.054571817e-34  # J·s (Constante de Planck réduite)
c = 2.99792458e8        # m/s (Vitesse de la lumière)
eps0 = 8.854187817e-12  # F/m (Permittivité du vide)
pi = np.pi
h = 2 * pi * hbar       # J·s (Constante de Planck)

# Données spécifiques à la transition 5S1/2 → 5P3/2 du Rubidium (D2)
lambda_transition = 780e-9       # m (Longueur d'onde de la transition)
omega_eg = 2 * pi * c / lambda_transition  # rad/s (Fréquence angulaire de la transition)

# Moment dipolaire électrique pour la transition D2 du Rubidium (Rb D2)
# Correspond à environ 4.227 ea0
d_eg = 3.58424e-29         # C·m (Valeur plus précise tirée de 4.227 ea0)

# Calcul de Gamma (largeur naturelle ou taux de désexcitation spontanée)
# Formule: Gamma = (omega_eg^3 * d_eg^2) / (3 * pi * eps0 * hbar * c^3)
Gamma = (omega_eg**3 * d_eg**2) / (3 * pi * eps0 * hbar * c**3)

# Calcul de l'intensité de saturation (I_sat)
# Formule standard: I_sat = (pi * h * c * Gamma) / (3 * lambda_transition**3)
# Note: Cette formule est couramment utilisée pour un atome à deux niveaux
# et elle est robuste pour correspondre à la valeur expérimentale pour le Rb D2.
I_sat = (pi * h * c * Gamma) / (3 * lambda_transition**3)

# Affichage des résultats
print(f"Moment dipolaire utilisé (d_eg) = {d_eg:.3e} C·m")
print(f"Gamma (largeur naturelle) = {Gamma/(2*pi)/1.e6:.3f} MHz")
print(f"I_sat (intensité de saturation) = {I_sat:.3f} W/m²")
print(f"I_sat (intensité de saturation) = {I_sat * 1e-1:.3f} mW/cm²") # conversion W/m² → mW/cm²
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes physiques
c = 3e8  # vitesse de la lumière [m/s]
hbar = 1.0545718e-34  # constante de Planck réduite [J.s]
kB = 1.380649e-23  # constante de Boltzmann [J/K]
pi = np.pi
h = 2 * pi * hbar

# --- PARAMÈTRES SPÉCIFIQUES AU RUBIDIUM D2 (Rb87) ---
# Gamma (taux de désexcitation spontanée) en rad/s
#Gamma = 2 * pi * 6.069e6
# Intensité de saturation en W/m^2
#Isat = 1.67

# Longueurs d'onde des transitions [m]
lambda_D1 = 795e-9
lambda_D2 = 780e-9
omega_D1 = 2 * pi * c / lambda_D1
omega_D2 = 2 * pi * c / lambda_D2

# Laser dans le bleu : plage de lambda de 755 à 779 nm
lambda_L = np.linspace(770e-9, 781e-9, 500)
omega_L = 2 * pi * c / lambda_L

# Désaccords
Delta1 = (omega_L - omega_D1)#*1.e200
Delta2 = (omega_L - omega_D2)#*2/3

# Coefficients pour le potentiel dipolaire et le taux de diffusion
coeff_dip = (1 / (3 * Delta1)) + (2 / (3 * Delta2))
coeff_sc = (1 / (3 * Delta1**2)) + (2 / (3 * Delta2**2))

# Calcul de eta (figure de mérite)
eta = (coeff_dip) / ( Gamma * coeff_sc)

# --- Définition des critères pour la puissance minimale ---
U_dip_min = kB * 1e-6         # Potentiel minimum souhaité (en Joule, pour 1 micro-Kelvin)
Gamma_sc_max = 2 * pi * 1 /(2*pi)     # Taux de diffusion maximum souhaité (en rad/s, pour 1 Hz)
waist = 1e-6                # Waist du faisceau Gaussien [m]

eta_min = U_dip_min / (hbar * Gamma_sc_max)/2

# Calcul de la puissance requise pour atteindre U_dip_min à chaque lambda_L
P_req_for_Udip_min = (4 * pi * I_sat * waist**2 * U_dip_min) / (hbar * Gamma**2 * coeff_dip)

# Calcul de la puissance maximale autorisée pour respecter Gamma_sc_max à chaque lambda_L
P_max_for_Gamma_sc_max = (8 * pi * I_sat * waist**2 * Gamma_sc_max) / (Gamma**3 * coeff_sc)

# --- Recherche de la puissance minimale et de la lambda optimale ---
valid_indices = np.where(eta >= eta_min)
valid_lambda_L = lambda_L[valid_indices]
valid_P_req_for_Udip_min = P_req_for_Udip_min[valid_indices]
valid_P_max_for_Gamma_sc_max = P_max_for_Gamma_sc_max[valid_indices]

P_min_overall = None
lambda_optimal_for_P_min = None

if len(valid_lambda_L) > 0:
    P_min_overall = np.min(valid_P_req_for_Udip_min)
    idx_P_min_overall = np.argmin(valid_P_req_for_Udip_min)
    lambda_optimal_for_P_min = valid_lambda_L[idx_P_min_overall]

    print(f"\n--- Résultats du Calcul de Puissance Optimale ---")
    print(f"Critères : U_dip_min = {U_dip_min/kB*1e6:.2f} µK, Gamma_sc_max = {Gamma_sc_max/(2*pi):.2f} Hz, waist = {waist*1e6:.0f} µm")
    print(f"Figure de mérite minimale requise (eta_min) : {eta_min:.2e}")
    print(f"La puissance minimale requise pour ces critères est de : {P_min_overall*1e3:.3f} mW")
    print(f"Cette puissance est atteinte à la longueur d'onde optimale : {lambda_optimal_for_P_min * 1e9:.3f} nm")
else:
    print(f"\n--- Résultats du Calcul de Puissance Optimale ---")
    print(f"eta_min = {eta_min:.2e}")
    print("Aucune combinaison de U_dip_min et Gamma_sc_max n'est réalisable dans la plage de longueurs d'onde spécifiée (755-779 nm).")

# --- PARAMÈTRE D'ERREUR ---
error_percentage = 60 # Pourcentage d'erreur
P_tolerance_limit = P_min_overall * (1 + error_percentage / 100) # Limite supérieure de puissance tolérée

# --- Tracer la puissance requise en fonction de la longueur d'onde ---
plt.figure(figsize=(10, 6))

plt.plot(lambda_L * 1e9, P_req_for_Udip_min * 1e3, color='blue', label=rf"$P$ requis pour $U_{{\mathrm{{dip}}}}(0) = U_{{\mathrm{{dip}}}}^{{\mathrm{{(min)}}}} = k_B \times ${U_dip_min/kB/1.e-6:.1f} $\mu$K" )
plt.plot(lambda_L * 1e9, P_max_for_Gamma_sc_max * 1e3, color='orange', linestyle='--', label=rf"$P$ requis pour $\Gamma_{{\mathrm{{sc}}}}(0) = \Gamma_{{\mathrm{{sc}}}}^{{\mathrm{{(max)}}}} =  2\pi \times ${Gamma_sc_max*2:.1f} s$^{{-1}}$")

# Marquer le point de puissance minimale optimale
if P_min_overall is not None:
    plt.axvline(x=lambda_optimal_for_P_min * 1e9, color='purple', linestyle=':')#,
                #label=rf"$\lambda_{{\mathrm{{max}}}}$ = {lambda_optimal_for_P_min * 1e9:.4f} nm")
    # Hachures à droite de la ligne
    plt.axvspan(lambda_optimal_for_P_min * 1e9, 781 , color='purple', alpha=0.1, hatch='///')

    plt.plot(lambda_optimal_for_P_min * 1e9, P_min_overall * 1e3, 'x', color='red', markersize=8,
             label=rf"$\lambda^{{\mathrm{{(min)}}}}_{{\mathrm{{abs}}}}$ = {lambda_optimal_for_P_min * 1e9:.1f} nm , P$^{{\mathrm{{(min)}}}}_{{\mathrm{{abs}}}}$ = {P_min_overall*1e3:.4f} mW")

    # --- Zone de tolérance à 60% d'erreur ---
    # Nous voulons colorer la zone où P_req_for_Udip_min <= P_tolerance_limit
    # ET P_req_for_Udip_min <= P_max_for_Gamma_sc_max
    # ET eta >= eta_min (déjà implicite par le calcul de P_req_for_Udip_min)

    # Trouver les indices où la puissance requise est dans la limite de tolérance et respecte Gamma_sc_max
    # On travaille avec les puissances en mW pour le seuil
    P_req_mW = P_req_for_Udip_min * 1e3
    P_max_sc_mW = P_max_for_Gamma_sc_max * 1e3
    P_tol_mW = P_tolerance_limit * 1e3

    # Conditions pour la zone colorée:
    # 1. P_req_mW <= P_tol_mW (puissance dans la tolérance)
    # 2. P_req_mW <= P_max_sc_mW (respecte la contrainte de scattering)
    # 3. eta >= eta_min (respecte la figure de mérite minimale)

    # On prend la zone où P_req est inférieur ou égal à la limite de tolérance
    # et aussi inférieur ou égal à la limite max de scattering
    #fill_indices = np.where((P_req_mW <= P_tol_mW) & (P_req_mW <= P_max_sc_mW) & (eta >= eta_min))
    fill_indices = np.where((P_req_for_Udip_min <= P_max_for_Gamma_sc_max) & (eta >= eta_min))

    if len(fill_indices[0]) > 0:
        # Trouver les limites de la zone colorée en longueur d'onde
        lambda_start = lambda_L[fill_indices[0][0]] * 1e9
        lambda_end = lambda_L[fill_indices[0][-1]] * 1e9

        # Plotter la zone colorée
        #plt.fill_between(lambda_L[fill_indices] * 1e9, P_req_for_Udip_min[fill_indices]* 1e3* (1+ 0.6 ), P_max_for_Gamma_sc_max[fill_indices]* 1e3,
                         #color='lightgreen', alpha=0.4, label=f'Zone tolérée (P < {P_tol_mW:.1f} mW)')
        #plt.fill_between(lambda_L[fill_indices] * 1e9, P_req_for_Udip_min[fill_indices]* 1e3* (1+ 0.6 ), P_max_for_Gamma_sc_max[fill_indices]* 1e3,
                         #color='lightgreen', alpha=0.4, label=f'Zone tolérée')
        plt.fill_between(lambda_L[fill_indices] * 1e9, P_req_for_Udip_min[fill_indices]* 1e3/(1- 0.75), 230 ,
                         color='lightgreen', alpha=0.4, label=f'Zone tolérée')

        # Ajouter une ligne pour la limite supérieure de puissance tolérée
        #plt.axhline(y=P_tol_mW, color='grey', linestyle=':', label=f'P$_{{\mathrm{{tol}}}}$ = {P_tol_mW:.1f} mW')

Laser = {
    "DFB-772.5": {"lambda": 772.5, "Puissance": 75},
    "DBR-778.0": {"lambda": 778.0, "Puissance": 60},
    "DBR-770.5": {"lambda": 770.5, "Puissance": 32}
}
# Couleurs pour les courbes
colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']

# Tracé automatique des lasers
for i, (name, params) in enumerate(Laser.items()):
    lam = params["lambda"]
    power = params["Puissance"]
    color = colors[i % len(colors)]

    P_required = np.interp(lam, lambda_L* 1e9, P_req_for_Udip_min * 1e3 /(1- 0.75))

    plt.plot(lam, power, 'o', color=color, markersize=8,
             label=rf"{name} : $\lambda$ = {lam:.1f} nm, $P$ = {power:.1f} mW")

    print(f"{name} ({lam} nm) : Puissance requise ≈ {P_required:.4f} mW, disponible = {power:.1f} mW")
    
# Tracé
#plt.plot(Laser["DFB"]["lambda"], Laser["DFB"]["Puissance"], 'o', color='blue', markersize=8,
         #label=rf"DFB : $\lambda =$ {Laser["DFB"]["lambda"]:.1f} nm, $P$ = {Laser["DFB"]["Puissance"]:.1f} mW")

#plt.plot(Laser["DBR"]["lambda"], Laser["DBR"]["Puissance"], 'o', color='green', markersize=8,
         #label=rf"DBR : $\lambda =$ {Laser["DBR"]["lambda"]:.1f} nm, $P$ = {Laser["DBR"]["Puissance"]:.1f} mW")

from matplotlib.ticker import ScalarFormatter

plt.xlabel("Longueur d'onde du laser $\lambda$ (nm)")
plt.ylabel("Puissance $P$ (mW)")
plt.title("Puissance requise vs. Longueur d'onde avec zone de tolérance (Rb87)")

plt.grid(True, which='both')  # grille sur les majeures et mineures
plt.legend()

# Limites des axes
plt.xlim(770, 780)
plt.ylim(0.00001, 120)  # éviter 0 pour une échelle log

# Échelle logarithmique pour y
plt.yscale("log")

# Affichage en puissance de 10
formatter = ScalarFormatter(useMathText=True)
formatter.set_powerlimits((0, 0))
plt.gca().yaxis.set_major_formatter(formatter)

plt.tight_layout()

# Sauvegarde avec haute résolution
plt.savefig("puissance_vs_lambda_log.png", dpi=300, bbox_inches='tight')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Paramètres initiaux
p0_V, p1_V = 2.0402672878862194, 0.38024521824423774
p0_TA, p1_TA, p2_TA = 382.9652981651816, 879.109917227037, +479.33716504028456
p0_TA, p1_TA, p2_TA = 0 , -175.8 , 441.5   
# Fonctions modèles
def fonc_V(x): return p0_V + p1_V * x
def fonc_TA(x): return p0_TA + p1_TA * x + p2_TA * x**2

# Bruit simulé (ajustable)
noise_level_V = 0.005  # écart-type du bruit pour V
noise_level_TA = 10.0  # écart-type du bruit pour TA

# Données X
X = np.array([0, 0.1, 0.3, 0.4, 0.55, 0.75, 0.9, 1.02, 1.2, 1.5, 1.75, 2.0, 2.1])

# Génération des données bruitées
V = fonc_V(X) + np.random.normal(0, noise_level_V, size=X.shape)
TA = fonc_TA(X) + np.random.normal(0, noise_level_TA, size=X.shape)
for i , ta in enumerate(TA):
    print(i, ta)
    if i == 0 : 
        while np.any(ta < 0): 
            ta = fonc_TA(X[i]) + np.random.normal(0, noise_level_TA, size=1)
    else : 
        while ta < TA[i-1] : 
            ta = fonc_TA(X[i]) + np.random.normal(0, noise_level_TA, size=1)
    print(i, ta)
    TA[i] = ta
while np.any(TA < 0):
    TA = fonc_TA(X) + np.random.normal(0, noise_level_TA, size=X.shape)

# Fit des données bruitées
def model_V(x, a, b): return a + b * x
def model_TA(x, a, b, c): return a + b * x + c * x**2

popt_V, _ = curve_fit(model_V, X, V)
popt_TA, _ = curve_fit(model_TA, X, TA)

# Affichage des résultats
print("Fit V(x) : a = {:.4f}, b = {:.4f}".format(*popt_V))
print("Fit TA(x) : a = {:.2f}, b = {:.2f}, c = {:.2f}".format(*popt_TA))

# Tracés
x_plot = np.linspace(0, 2.2, 200)

plt.figure(figsize=(12, 5))

# Graphique pour V
plt.subplot(1, 2, 1)
plt.scatter(X, V, label="Points bruités", color="blue")
plt.plot(x_plot, fonc_V(x_plot), '--', label="Modèle initial", color="gray")
plt.plot(x_plot, model_V(x_plot, *popt_V), label="Fit", color="red")
plt.title("Fit de V(x)")
plt.xlabel("x")
plt.ylabel("V")
plt.legend()

# Graphique pour TA
plt.subplot(1, 2, 2)
plt.scatter(X, TA, label="Points bruités", color="green")
plt.plot(x_plot, fonc_TA(x_plot), '--', label="Modèle initial", color="gray")
plt.plot(x_plot, model_TA(x_plot, *popt_TA), label="Fit", color="red")
plt.title("Fit de TA(x)")
plt.xlabel("x")
plt.ylabel("TA")
plt.legend()

plt.tight_layout()
plt.show()
```
